```python
import matplotlib.pyplot as plt
from matplotlib import mlab
%matplotlib inline
import numpy as np

sr = 20000
freq1 = 200
freq2 = 100

x = np.linspace(0, 1, sr)
y = np.concatenate([0.5*np.sin(freq1 * 2 * np.pi * x[:10000]) + 0.5*np.sin(freq2 * 2 * np.pi * x[:10000]), np.sin(freq1 * 2 * np.pi * x[10000:15000]), np.sin(freq2 * 2 * np.pi * x[15000:20000])])

fig, ax = plt.subplots(2, 1, figsize=(8, 4))

ax[0].set_title('(a)')
ax[0].plot(x, y)
ax[0].axis(xmin=0, xmax=1)
ax[0].set_xlabel('Čas [s]')
ax[0].set_ylabel('Amplituda [-]')

ax[1].set_title('(b)')
ax[1].magnitude_spectrum(y, Fs=sr, color='C1', window=mlab.window_none)
ax[1].axis(xmin=0, xmax=500)
ax[1].set_xlabel('Frekvence [Hz]')
ax[1].set_ylabel('Amplituda [-]')

plt.tight_layout()

plt.savefig('casova_a_frekvencni_oblast.pdf', format='pdf')

plt.show()
```

Tato buňka generuje obrázek znázorňující zobrazení signálu z předchozího obrázku v časově frekvenční oblasti. <br>
Dále je je z obrázku patrný vliv délky okna na výsledek frekvenční analýzy.

```python
import matplotlib.pyplot as plt
from matplotlib import mlab
%matplotlib inline
import numpy as np
import librosa
import librosa.display
import matplotlib.ticker as ticker

def fmt(x, pos):
    res = int(x)/sr
    return '{}'.format(res)

sr = 20000
freq1 = 200
freq2 = 100
n_fft1 = 1000
n_fft2 = 2000

x = np.linspace(0, 1, sr)
y = np.concatenate([0.5*np.sin(freq1 * 2 * np.pi * x[:10000]) + 0.5*np.sin(freq2 * 2 * np.pi * x[:10000]), np.sin(freq1 * 2 * np.pi * x[10000:15000]), np.sin(freq2 * 2 * np.pi * x[15000:20000])])

no_window1 = np.linspace(1, 1, n_fft1)
D1 = np.abs(librosa.stft(y, n_fft=n_fft1, hop_length=int(n_fft1/2), window=no_window1, center=False))

fig, ax = plt.subplots(2, 1, figsize=(9, 4), sharey=True)

librosa.display.specshow(D1, y_axis='linear', ax=ax[0], sr=sr, hop_length=int(n_fft1/2))

cbar = plt.colorbar(ax[0].collections[0], ax=ax[0], format=ticker.FuncFormatter(fmt))
cbar.set_label('Amplituda [-]')

ax[0].set_title('(a)')
# ax[0].set_xlabel('Čas [s]')
ax[0].set_ylabel('Frekvence [Hz]')
ax[0].axis(ymax=500)
ax[0].set_xticks([0, 7, 15, 23, 31, 39])
ax[0].set_xticklabels(['' for i in [0, 7, 15, 23, 31, 39]])

no_window2 = np.linspace(1, 1, n_fft2)
D2 = np.abs(librosa.stft(y, n_fft=n_fft2, hop_length=int(n_fft2/2), window=no_window2, center=False))

librosa.display.specshow(D2, y_axis='linear', ax=ax[1], sr=sr, hop_length=int(n_fft1/2))

cbar = plt.colorbar(ax[1].collections[0], ax=ax[1], format=ticker.FuncFormatter(fmt))
cbar.set_label('Amplituda [-]')

ax[1].set_title('(b)')
ax[1].set_xlabel('Čas [s]')
ax[1].set_ylabel('Frekvence [Hz]')
ax[1].axis(ymax=500)
ax[1].set_xticks([0, 3, 7, 11, 15, 19])
ax[1].set_xticklabels(['{:.1f}'.format(i/19) for i in [0, 3, 7, 11, 15, 19]])

plt.tight_layout()

plt.savefig('spektrogramy.pdf', format='pdf')

plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.datasets import make_blobs


# we create 40 separable points
X1, y1 = make_blobs(n_samples=40, centers=1, random_state=7, cluster_std=1)
X2, y2 = make_blobs(n_samples=40, centers=1, random_state=5, cluster_std=1)

fig, ax = plt.subplots(1, 2, figsize=(8, 4))

point = [-7.5, 7]

ax[0].scatter(X1[:, 0], X1[:, 1], color='r', marker='^', s=10**2, zorder=10)
ax[0].scatter(X2[:, 0], X2[:, 1], color='b', marker='o', s=10**2, zorder=10)
ax[0].plot(point[0], point[1], 'gs', markersize=10, zorder=10)

# plot the decision function

ax[0].axis(xmin=-9, xmax=-5, ymin=5, ymax=9)
ax[0].set_xticks(np.arange(-9, -4.5, 0.5))
ax[0].set_yticks(np.arange(5, 9.5, 0.5))
ax[0].set_aspect("equal")
xlim = ax[0].get_xlim()
ylim = ax[0].get_ylim()

X = np.concatenate((X1, X2))

knn = NearestNeighbors(n_neighbors=5)
knn.fit(X)

pts = knn.kneighbors([point], return_distance=False)

for pt in pts[0]:
    ax[0].plot([point[0], X[pt][0]], [point[1], X[pt][1]], 'k--', zorder=0)

ax[0].set_yticklabels([])
ax[0].set_xticklabels([])
ax[0].set_title('(a)')

# we create 40 separable points
X1, y1 = make_blobs(n_samples=40, centers=1, random_state=7, cluster_std=1)
X2, y2 = make_blobs(n_samples=40, centers=1, random_state=5, cluster_std=1)

point = [-7.5, 7]

ax[1].scatter(X1[:, 0], X1[:, 1], color='r', marker='^', s=10**2, zorder=10)
ax[1].scatter(X2[:, 0], X2[:, 1], color='b', marker='o', s=10**2, zorder=10)
ax[1].plot(point[0], point[1], 'gs', markersize=10, zorder=10)

# plot the decision function

ax[1].axis(xmin=-9, xmax=-5, ymin=5, ymax=9)
ax[1].set_xticks(np.arange(-9, -4.5, 0.5))
ax[1].set_yticks(np.arange(5, 9.5, 0.5))
ax[1].set_aspect("equal")
xlim = ax[1].get_xlim()
ylim = ax[1].get_ylim()

X = np.concatenate((X1, X2))

knn = NearestNeighbors(n_neighbors=7)
knn.fit(X)

pts = knn.kneighbors([point], return_distance=False)

for pt in pts[0]:
    ax[1].plot([point[0], X[pt][0]], [point[1], X[pt][1]], 'k--', zorder=0)

ax[1].set_yticklabels([])
ax[1].set_xticklabels([])
ax[1].set_title('(b)')
    
plt.tight_layout()
plt.savefig('knn.pdf', format='pdf')

plt.show()
```

Tato buňka generuje obrázek znázorňující princip fungování metody podpůrných vektorů s a bez použití jádrové metody.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs, make_circles


# we create 40 separable points
X, y = make_blobs(n_samples=40, centers=2, random_state=7, cluster_std=3)

# fit the model, don't regularize for illustration purposes
clf = svm.SVC(kernel='linear', C=1000)
clf.fit(X, y)

fig, ax = plt.subplots(1, 2, figsize=(8, 4))

ax[0].scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax[0].axis(xmin=-11, xmax=3, ymin=-2, ymax=12)
ax[0].set_aspect("equal")
xlim = ax[0].get_xlim()
ylim = ax[0].get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax[0].contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'], zorder=0)

# plot support vectors
ax[0].scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=200, linewidth=1, facecolors='none', edgecolors='k')

ax[0].set_yticklabels([])
ax[0].set_xticklabels([])
ax[0].set_title('(a)')

X, y = make_circles(80, factor=.1, noise=.2, random_state=6)

clf = svm.SVC(kernel='rbf', C=1000)
clf.fit(X, y)

ax[1].scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax[1].axis(xmin=-1.5, xmax=1.5, ymin=-1.5, ymax=1.5)
ax[1].set_aspect("equal")
xlim = ax[1].get_xlim()
ylim = ax[1].get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax[1].contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'], zorder=0)

# plot support vectors
ax[1].scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=200, linewidth=1, facecolors='none', edgecolors='k')

ax[1].set_yticklabels([])
ax[1].set_xticklabels([])
ax[1].set_title('(b)')

plt.tight_layout()
plt.savefig('svm.pdf', format='pdf')

plt.show()
```
