```python
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from z_hpoptimise import HPoptimise
```

```python
def plot(y_test, y_pred):
    fig, ax = plt.subplots(5, 1, figsize=(18, 18))
    ax[0].scatter(y_test[:,0], y_test[:,1], label='True values', color='blue')
    ax[0].scatter(y_pred[:,0], y_pred[:,1], label='Predicted values', color='red')
    ax[0].set_title('Point in 2d space')
    ax[0].legend()

    ticks = np.linspace(1, len(y_test), num=len(y_test))

    ax[1].scatter(ticks, y_test[:,0], label='True values', color='blue')
    ax[1].scatter(ticks, y_pred[:,0], label='Predicted values', color='red')
    ax[1].set_title('A')
    ax[1].legend()

    ax[2].scatter(ticks, y_test[:,1], label='True values', color='blue')
    ax[2].scatter(ticks, y_pred[:,1], label='Predicted values', color='red')
    ax[2].set_title('B')
    ax[2].legend()

    diff1 = np.abs(y_test[:,0] - y_pred[:,0])

    ax[3].scatter(ticks, diff1, label='', color='blue')
    ax[3].set_title('abs(A_true - A_pred)')

    diff2 = np.abs(y_test[:,1] - y_pred[:,1])

    ax[4].scatter(ticks, diff2, label='', color='blue')
    ax[4].set_title('abs(B_true - B_pred)')

    plt.tight_layout()
    plt.show()
```

```python
X = pd.read_csv('Coords.data' , sep=" ", index_col=0, header=None)
y = pd.read_csv('AH.data' , sep=" ", index_col=0, header=None)

X = X.values
y = y.values

print('Correlation coefficient of A and H: {}'.format(np.corrcoef(y[:,0], y[:,1])[0][1]))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)

# model = RandomForestRegressor

# estim = HPoptimise(X_train=X_train, y_train=y_train, X_val=X_val, y_val=y_val, Regressor=model, filepath='study.pkl', max_trials=5000, max_trials_no_change=100)
# estim.optimise()

# params = estim.best_params()

model = RandomForestRegressor()

model.fit(X_train, y_train)
print(model)
y_pred = model.predict(X_test)

print('R2 score uniform average: {}'.format(r2_score(y_test, y_pred)))

plot(y_test, y_pred)
```
